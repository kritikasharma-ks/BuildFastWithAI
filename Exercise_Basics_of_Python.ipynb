###Hint

###8. Working with Libraries

###Exercise 1

Use Matplotlib to create a line plot of the following data:
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
Label the x-axis as "X-axis" and the y-axis as "Y-axis". Add a title to the plot.



```python
# Logic: Use Matplotlib to create a line plot
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
```

```python
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.show()
```

###Exercise 2

Create a Pandas DataFrame with the following data:
Name: ["Alice", "Bob", "Charlie"]
Age: [25, 30, 35]
City: ["New York", "San Francisco", "Los Angeles"]
Print the DataFrame and the mean age.

```python
# Import the Pandas library
import pandas as pd
# Create a dictionary with the given data
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "San Francisco", "Los Angeles"]
}

# Logic: Convert the dictionary into a Pandas DataFrame

df = pd.DataFrame(data)
```

```python
print(df)
print("\nMean age: ", df["Age"].mean())
```
