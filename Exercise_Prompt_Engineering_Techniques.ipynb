###Exercise 1: Few-Shot Learning




Create a few-shot prompt that generates a multiple-choice question about a given topic. Use the GPT-4o model to generate a question about the topic "Artificial Intelligence."

####Hint

```python
from langchain_openai import ChatOpenAI
from langchain_google_genai import ChatGoogleGenerativeAI
from google.colab import userdata
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
gpt4o_mini_model = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY)

GEMINI_API_KEY = userdata.get('GEMINI_API_KEY')
gemini_model = ChatGoogleGenerativeAI(model="gemini-2.5-flash", api_key=GEMINI_API_KEY)
```

```python
# Example: User wants a multiple-choice question about Artificial Intelligence.
input_topic = "Artificial Intelligence"

# Logic: Use a few-shot prompt to generate a multiple-choice question.
# Hint: Initialize the prompt with examples and use `invoke` to send the request.

few_shot_prompt = """
Give a multiple choice question about {input_topic}. The format should be the following:

Question 1: What is the capital of France?
A. Paris
B. London
C. Berlin
D. Madrid

Answer: A. Paris

Question 2: What is the largest planet in our solar system?
A. Mars
B. Jupiter
C. Saturn
D. Earth

Answer: B. Jupiter
"""

response = gpt4o_mini_model.invoke(few_shot_prompt.format(input_topic=input_topic))

print(response.content)




```

###Exercise 4: Chain-of-Thought Prompting



Use chain-of-thought prompting to solve a math problem step by step. The problem is: "What is the result of 45 x 12 + 18 ÷ 6 - 100?"

```python
# Example: User wants to solve a math problem step by step.
math_problem = "What is the result of 45 x 12 + 18 ÷ 6 - 100?"

math_prompt = """

Let's solve arithmetic problems step by step. For the given problem: {math_problem}.

- Identify the operations in the problem (addition, subtraction, multiplication, division).
- Recall the order of operations: PEMDAS (Parentheses, Exponents, Multiplication and Division from left to right, Addition and Subtraction from left to right).
- Perform the operations in the correct order, showing your work at each step.
- Double-check your calculation.

Provide the final answer.
"""

# Logic: Use chain-of-thought prompting to solve the problem.
# Hint: Initialize the prompt with step-by-step instructions and use `invoke` to send the request.

response = gpt4o_mini_model.invoke(math_prompt.format(math_problem=math_problem))

print(response.content)
```
